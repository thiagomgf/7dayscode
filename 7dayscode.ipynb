# **Pré-processamento dos Dados**

Vou começar com a indicação da criação da coluna informando se a música é popular ou não, seguindo a orientação de considerar uma popularidade >70 como parâmetro de corte.

```python
# 1. Definindo a condição de popularidade (maior que 70)
# 2. Convertendo a condição booleana diretamente para inteiro (True se torna 1, False se torna 0)
dados_musicas['is_popular'] = (dados_musicas['popularity'] > 70).astype(int)

# 3. Exibindo as primeiras linhas para verificar a nova coluna
dados_musicas.head()
```

```python
# Filtrando apenas as músicas populares (is_popular == 1)
musicas_populares = dados_musicas[dados_musicas['is_popular'] == 1]

# Agrupando por artista para contar as músicas populares e calcular a média de popularidade
artistas_pop_stats = musicas_populares.groupby('artists').agg(
    popular_songs_count=('is_popular', 'sum'), # Contagem de músicas populares
    avg_popularity=('popularity', 'mean')      # Média da popularidade
).sort_values(by='popular_songs_count', ascending=False).reset_index()

# Exibindo os 5 principais artistas com mais músicas populares
top_5_artistas_populares = artistas_pop_stats.head(5)
top_5_artistas_populares
```
